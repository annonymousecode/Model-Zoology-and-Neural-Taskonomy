```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os, pickle
```

```python
brain_rdms = pickle.load(open("brain_rdms_reliable.pkl", "rb" ))
```

```python
brain_rdms_dict = {}
for area in brain_rdms:
    for layer in brain_rdms[area]:
        brain_rdms_dict[area + '_' + layer] = brain_rdms[area][layer][np.triu_indices(119,k=1)]
        
brain_rdm_df = pd.DataFrame(brain_rdms_dict)
output_file = 'brain_uber_rdm.csv'
if not os.path.exists(output_file):
    brain_rdm_df.to_csv(output_file, index = None)
```
